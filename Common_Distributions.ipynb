# Common Probability Distributions

This notebook explores **discrete and continuous probability distributions**, their properties, and visualizations.

For each distribution:

```python
#import libraries 

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from statsmodels import *
from scipy import *
```

## **Discrete Variable Distributions**

### **Bernoulli Distribution**
A **Bernoulli trial** is a single experiment with **two possible outcomes**:
- Success (\(1\)): Heads in a coin flip
- Failure (\(0\)): Tails in a coin flip

#### **ðŸ”¹ Real-World Examples**
- **Medical Trials:** A drug is either effective (1) or not (0).
- **Spam Detection:** An email is either spam (1) or not spam (0).



```python
##define parameters

#probability of success
p=0.5
#define possible outcomes in list
x=[0,1]
#get values of pmf 
pmf_values = stats.bernoulli.pmf(x,p)

```

```python
#plot the pmf 
plt.bar(x, pmf_values, color='blue', alpha=0.7)
plt.xticks([0, 1], labels=["Tails (0)", "Heads (1)"])
plt.ylabel("Probability")
plt.title("Bernoulli Distribution PMF (Coin Flip)")
plt.show()
```

```python
#plot the cdf
cdf_values = stats.bernoulli.cdf(x,p)

#plot the pmf 
plt.step(x, cdf_values, color='blue', alpha=0.7)
plt.xticks([0, 1], labels=["Tails (0)", "Heads (1)"])
plt.ylabel("Cumulative Probability")
plt.title("Bernoulli CDF")
```

### **Binomial Distribution**
A **binomial distribution** is the sum of **multiple Bernoulli trials**.

#### **ðŸ”¹ Real-World Examples**
- **Manufacturing:** A factory produces light bulbs with a **5% defect rate**. If we test **100 bulbs**, how many are defective?
- **Marketing:** If **20% of users** click an ad, how many clicks will we get from **50 visitors**?

```python
#define parameters
n, p= 10,0.5 

#define possible outcomes 
x=np.arange(0,n+1)

#define pmf values
pmf_values_bin=stats.binom.pmf(x,n,p)
```

```python
#plot the pmf 

plt.bar(x,pmf_values_bin,color='purple',alpha=0.7)
plt.xlabel("Number of Heads")
plt.ylabel("Probability")
plt.title("Binomial Distribution (PMF) of Coin Flips")
plt.show()
```

```python
#define variable to hold cdf values
cdf_values_bin = stats.binom.cdf(x,n,p)

#plot the cdf 
plt.step(x,cdf_values_bin,color='purple',alpha=0.7)
plt.xlabel("Number of Heads")
plt.ylabel("Probability")
plt.title("Binomial Distribution (CDF) of Coin Flips")
plt.show()
```

### **Poisson Distribution**
The **Poisson distribution** models the probability of **a given number of events occurring in a fixed time period**, assuming that these events happen independently and at a constant average rate.

#### **ðŸ”¹ Real-World Examples**
- **Customer Arrivals:** Number of customers arriving at a store per hour.
- **Flight Passengers:** The number of passengers arriving at an airport in a day.
- **Emails Received:** Number of emails received per hour.
- **Calls at a Call Center:** Incoming call volume per minute.
- **Biology:** Mutations in DNA over time.


```python
#define parameter labmda in this instance, choosing random for purpose of practice
lambda_p=5
#variable to hold possible range of passengers that can be on the flight
x_p = np.arange(0,24,2)
#define pmf values
pmf_values_p=stats.poisson.pmf(x_p,lambda_p)
```

```python
#plot pmf 
plt.bar(x_p,pmf_values_p,color="green",alpha=0.7)
plt.xlabel("Number of Passengers")
plt.ylabel("Probability")
plt.title("Poisson Distribution (PMF) Flight Passengers")
plt.show()
```

```python
#define cdf values 
cdf_values_p=stats.poisson.cdf(x_p,lambda_p)

#plot cdf
plt.step(x_p,cdf_values_p,color="green",alpha=0.7)
plt.xlabel("Number of Passengers")
plt.ylabel("Cumulative Probability")
plt.title("Poisson Distribution (PMF) Flight Passengers")
plt.show()
```

## **Continuous Variable Distributions**

### **Normal Distribution**
A **Normal (Gaussian) Distribution** is used for data that clusters around a mean.

#### **ðŸ”¹ Real-World Examples**
- **Height of people** follows a normal distribution.
- **SAT scores** are normally distributed with a mean of 1000 and a standard deviation of 200.
- **Stock market returns** are often modeled using a normal distribution.

```python
#define parameters
mu, sigma= 50, 10

#create a range to generate the distribution for
#arguments min,max,how many data points you want generated
norm_points=np.linspace(mu - 4*sigma, mu + 4*sigma,100)


#define pdf values
pdf_values_norm=stats.norm.pdf(norm_points,loc=mu,scale=sigma)
```

```python
#plot pdf 
plt.bar(norm_points,pdf_values_norm,color="green",alpha=0.7)
plt.title("Normal Distribution (PDF)")
plt.show()
```

```python
#define cdf values
cdf_values_norm=stats.norm.cdf(norm_points,loc=mu,scale=sigma)

#plot cdf
plt.bar(norm_points,cdf_values_norm,color="green",alpha=0.7)
plt.title("Normal Distribution (CDF)")
plt.show()
```

### **Exponential Distribution**
An **Exponential Distribution** models **the time between events** in a Poisson process.

#### **ðŸ”¹ Real-World Examples**
- **Waiting time for a bus** that arrives randomly every 10 minutes.
- **Time between earthquakes** in a region.
- **Customer arrivals at a store**.

```python
#define parameters
lambda_exp=2
one_over_lambda = 1 / lambda_exp


#define range of values 
expon_range = np.linspace(0,6,100)

#define pdf 
pdf_values_expon=stats.expon.pdf(expon_range,scale=one_over_lambda)
```

```python
#plot pdf 
plt.bar(expon_range,pdf_values_expon,color="black",alpha=0.7)
plt.title("Exponential Distribution (PDF)")
plt.show()
```

```python
#define cdf values
cdf_values_expon=stats.expon.cdf(expon_range,scale=one_over_lambda)

#plot cdf
plt.step(expon_range,cdf_values_expon,color="black",alpha=0.7)
plt.title("Exponential Distribution (CDF)")
plt.show()
```

### **Uniform Distribution**
The **Uniform distribution** describes a situation where **all outcomes are equally likely** within a given range.

#### **ðŸ”¹ Real-World Examples**
- **Rolling a Fair Die:** Each number (1-6) has an equal probability.
- **Random Number Generation:** Random selection between two numbers.
- **Lottery Draws:** Each ticket has an equal chance of winning.
- **Traffic Lights:** If a stoplight changes randomly within a range.

```python
#define parameters
a, b = 2, 10
uniform_range = b - a 

#generate range to create distribution over 
uni_values = np.linspace(a, b, 1000)

#get pdf values
pdf_values_uni = stats.uniform.pdf(uni_values,loc=a,scale=uniform_range)
```

```python
#plot pdf 
plt.bar(uniform_range,pdf_values_uni,color="black",alpha=0.7)
plt.title("Exponential Distribution (PDF)")
plt.show()
```

```python
#define cdf values
cdf_values_uni=stats.uniform.cdf(uni_values,loc=a,scale=uniform_range)

#plot cdf
plt.step(uni_values,cdf_values_uni,color="green",alpha=0.7)
plt.title("Uniform Distribution (CDF)")
plt.show()
```

## **Summary of Parametes for Each Distribution**


### ðŸ“Œ Discrete Probability Distributions (PMF & CDF)

| **Distribution**  | **PMF Function** | **CDF Function** | **Key Parameters** |
|------------------|------------------|-----------------|----------------|
| **Bernoulli**  | `stats.bernoulli.pmf(x, p)`  | `stats.bernoulli.cdf(x, p)`  | `p` â†’ Probability of success (0 â‰¤ p â‰¤ 1) |
| **Binomial**  | `stats.binom.pmf(x, n, p)`  | `stats.binom.cdf(x, n, p)`  | `n` â†’ Number of trials, `p` â†’ Probability of success |
| **Poisson**  | `stats.poisson.pmf(x, Î»)`  | `stats.poisson.cdf(x, Î»)`  | `Î»` â†’ Expected number of events per interval |

### ðŸ“Œ Key Formulas:
- **Bernoulli PMF:** $$ P(X=x) = p^x (1-p)^{1-x}, \quad x \in \{0,1\} $$
- **Binomial PMF:** $$ P(X=k) = \binom{n}{k} p^k (1-p)^{n-k} $$
- **Poisson PMF:** $$ P(X=k) = \frac{\lambda^k e^{-\lambda}}{k!} $$


### ðŸ“Œ Continuous Probability Distributions (PDF & CDF)

| **Distribution**  | **PDF Function** | **CDF Function** | **Key Parameters** |
|------------------|------------------|-----------------|----------------|
| **Normal (Gaussian)** | `stats.norm.pdf(x, loc=Î¼, scale=Ïƒ)` | `stats.norm.cdf(x, loc=Î¼, scale=Ïƒ)` | `Î¼` â†’ Mean, `Ïƒ` â†’ Standard deviation |
| **Exponential** | `stats.expon.pdf(x, scale=1/Î»)` | `stats.expon.cdf(x, scale=1/Î»)` | `Î»` â†’ Rate parameter (`scale = 1/Î»`) |
| **Uniform** | `stats.uniform.pdf(x, loc=a, scale=b-a)` | `stats.uniform.cdf(x, loc=a, scale=b-a)` | `a` â†’ Lower bound, `b` â†’ Upper bound (`scale = b-a`) |

### ðŸ“Œ Key Formulas:
- **Normal PDF:** $$ f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x-\mu)^2}{2\sigma^2}} $$
- **Exponential PDF:** $$ f(x) = \lambda e^{-\lambda x}, \quad x \geq 0 $$
- **Uniform PDF:** $$ f(x) = \frac{1}{b-a}, \quad a \leq x \leq b $$
